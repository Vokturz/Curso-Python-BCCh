```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("F29_DECLARANTES.xlsb", header=4)
# Necesitaras ejecutar !pip install pyxlsb
```

En pandas usar "~" significa negar el boolean, así "~True" será "False" y viceversa

Existen distintas operaciones que puedo hacer sobre una columna que del tipo string, todas ellas se encuentran en df["nombre_columna"].str

```python
# Limpiamos el dataframe
df_clean = df[~df["Tramo segun ventas (5 tramos)"].str.lower().str.contains("sin información")]
df_clean = df_clean[~df_clean["Región del domicilio o casa matriz"].str.lower().str.contains("sin información")]
df_clean = df_clean[~df_clean["Rubro económico"].str.lower().str.contains("sin información")]
```

```python
# Creamos la columna año
df_clean["Año"] = df_clean["Año Mes Comercial"].div(100).astype(int)
```

```python
# Miramos como quedó
df_clean
```

```python
# Graficar Año y total Número de declarantes según Medio de presentación
fig, ax = plt.subplots(figsize=(12,4))
sns.lineplot(data=df_clean, x="Año", y="Número de declarantes", hue="Medio de presentación", marker='o', estimator="sum", ax=ax)
plt.xticks(range(2005, 2023))
plt.grid()
plt.savefig("f29_anno_total_declarantes_medio.pdf")
plt.show()
```

```python
# Graficar Año y total Número de declarantes según Género asociado al RUT
fig, ax = plt.subplots(figsize=(12,4))
sns.lineplot(data=df_clean, x="Año", y="Número de declarantes", hue="Género asociado al RUT", marker='o', estimator="sum", ax=ax)
plt.xticks(range(2005, 2023))
plt.grid()
plt.savefig("f29_anno_total_declarantes_genero.pdf")
plt.show()
```

```python
# Podemos calcular la diferencia entre género masculino y femenino
resultado_genero = df_clean.groupby(["Año", "Género asociado al RUT"])["Número de declarantes"].sum().reset_index()
resultado_genero_F = resultado_genero[resultado_genero["Género asociado al RUT"] == "Femenino"]
resultado_genero_M = resultado_genero[resultado_genero["Género asociado al RUT"] == "Masculino"]
resultado_genero_diff = resultado_genero_F.set_index("Año")["Número de declarantes"] - resultado_genero_M.set_index("Año")["Número de declarantes"]
```

```python
# Graficamos la diferencia
resultado_genero_diff.plot(figsize=(10,3), xticks=range(2005, 2023, 2), marker=".")
plt.grid()
plt.title("Diferencia en número de\ndeclarantes Femenino y Masculino")
plt.ylabel("Diferencia en número de declarantes")
plt.savefig("f29_anno_total_declarantes_diff_genero.png", bbox_inches='tight', dpi=500)
plt.show()
```

```python
# Graficar Año y total Número de declarantes según Medio de presentación
fig, ax = plt.subplots(figsize=(12,4))
sns.lineplot(data=df_clean, x="Año", y="Número de declarantes",
             hue="Región del domicilio o casa matriz",
             marker='o', estimator="sum", ax=ax)
plt.xticks(range(2005, 2023))
plt.grid()
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5)) # para que el grafico esté fuera
plt.savefig("f29_anno_total_declarantes_region.pdf")
plt.semilogy() # modificar la escala de y a logaritmica
plt.show()
```
